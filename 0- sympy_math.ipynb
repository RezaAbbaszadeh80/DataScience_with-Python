```python
A = sy.Matrix([[1,2,3],[4,5,6]])
pprint(A)

a,b,c = sy.symbols('a,b,c')
B = sy.Matrix([[1,a,a**2],[1,b,b**2],[1,c,c**2]])  # ماتریس نمادین
pprint(B)

C = sy.Matrix([[-1,3],[2,1]])
pprint(C)

pprint(C*A)
## pprint(A*C) # ابعاد به هم نمیخوره و ارور میده

pprint(A*B)
## pprint(B*A)  # ابعاد به هم نمیخوره و ارور میده

D = sy.Matrix([[3,2],[-1,5]])

pprint(C+D)
pprint(D-C)

A = sy.Matrix([[2,-1,4],[3,-2,6],[8,1,5]])
pprint(A)
pprint(A.transpose())  # ترانهاده
pprint(A.adjoint()) # خود ترانهاده رو میده چون که مختلط نیست

B = sy.Matrix([[sy.I,1+2*sy.I],[-1,1-sy.I]])
pprint(B)
pprint(B.transpose())
pprint(B.adjoint())  # اول ترانهاده کرده بعدش مزدوج هرکدوم را حساب کرده

# تعریف ماتریس مربعی
A = sy.Matrix([[1, 2], [3, 4]])

pprint(sy.det(A))
pprint(A.inv())  # معکوس ماتریس
pprint(A.trace()) # مجموع عناصر روی قطر

A = sy.Matrix([[2,2,0],[2,1,1],[-7,2,-3]])
pprint(A)
pprint(A.eigenvals())  # محاسبه مقادیر ویژه

# کد محاسبه مقادیر ویژه به صورت دستی
c = sy.Symbol('c')
pprint(c*sy.eye(3) )  # در قطرش این حرف رو ضرب می کنه
B = A - c*sy.eye(3) 
p = B.det()
pprint(sy.solve(p,c)) 

pprint(A.eigenvects())  #بردار های ویژه
```
