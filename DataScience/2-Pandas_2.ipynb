# استفاده از عملگر های بولینی

```python
# سیاره‌هایی که دوره مداری آن‌ها نه دقیقاً یک سال است و نه کمتر از 200 روز
planets[~((planets['Orbital Period (days)'] == 365.25) | (planets['Orbital Period (days)'] < 200))] # ~  عملگر نقض
```

```python
# تفاوت عملگر نقض و نامساوی 
# نقض (~)
# این عملگر معمولاً برای نقض یا منفی کردن یک شرط یا مجموعه از شروط استفاده می‌شود
# به عبارت دیگر، اگر یک شرط صحیح باشد، نقض آن به صورت غلط خواهد بود و برعکس
# مثال: ~(x > 5) برابر با x <= 5 

# نامساوی (!=)
# این عملگر برای چک کردن نامساوی دو مقدار یا عبارت استفاده می‌شود
#  خواهد بود (True) اگر دو مقدار یکسان نباشند، عبارت نامساوی درست 
# مثال: x != 10 =====> برابر 10 نیست x یعنی 

# به عبارت دیگر، نقض یک شرط معکوس یک شرط است (اگر شرط صحیح باشد، نقض آن غلط خواهد بود و برعکس)
# در حالی که عملگر نامساوی مستقل از تعداد شروط یک مقدار یا عبارت را با مقدار دیگر مقایسه می‌کند
#=========================================================================================================

# سیاره‌هایی که دوره مداری آن‌ها یک سال نیست یا تعداد ماه‌های آن‌ها یک یا 2 است
selected_planets = planets[(planets['Orbital Period (days)'] != 365.25) | (planets['Number of Moons'].isin([1, 2]))]
selected_planets
```

# اعمال یک تابع روی دیتافریم

```python
# برای اعمال یک تابع روی جدول :
# 1- بر روی ستون یا سطرها  apply() تعریف تابع مورد نظر و  استفاده از تابع 
# 2- برای گروه‌بندی  groupby استفاده از تابع 
# 3- برای اعمال یک یا چند تابع تجمیعی به ستون‌ها agg استفاده از تابع 

# pd.NA    ==> هست یا باشد تبدیل شود به این  Nan 
# pd.notna ==> نیست یا نباشد یا تبدیل شود به این Nan 
# pd.NaT   ==> استفاده می شود NaT (Not a Time) معمولاً برای نمایش مقدار خالی در ستون‌های زمانی 

# planets.notna یا planets.notnull ==> برای بررسی عدم خالی بودن یک مقدار در یک سری یا دیتافریم و این دو تابع معادل هم هستند و مقدار بولینی برمی گردانند
# # این توابع برای فیلتر کردن داده‌ها بر اساس عدم خالی بودن یک ستون یا سری مفید هستند
```

```python
# در خود پایتون map() استفاده می کنیم مثل تابعapply() برای اعمال یک تابع بر روی یک دیتا فریم از متد 

def custom_sort(value):
    return value + 1400  # نباشند Nan با این مقدار جمع میشوند البته اون هایی که مقدار 

# 'Discovery Date' اعمال تابع سفارشی به ستون 
sorted_discovery_date = planets['Discovery Date'].apply(custom_sort)
sorted_discovery_date
```

```python
# تابع برای چاپ نزدیکی سیارات

def print_near_planets():
    # تبدیل مقادیر به اعداد صحیح
    planets['Orbital Radius (km)'] = planets['Orbital Radius (km)'].astype(float)
    
    # 'Orbital Radius (km)'مرتب کردن  بر اساس ستون 
    sorted_planets = planets.sort_values(by='Orbital Radius (km)')
    
    # چاپ نزدیکی سیارات
    for index, row in sorted_planets.iterrows():
        distance_int = int(row['Orbital Radius (km)'])
        
         # اگر مقدار دورانی کمتر یا مساوی 227900000 باشد، نام سیاره چاپ می‌شود
        if distance_int <= 227900000 and row['Name'] not in ['Sun', 'Earth']:
            print(f"{row['Name']} is near with Orbital Radius : {distance_int} ")

# اجرای تابع print_near_planets
print_near_planets()
```

# هدف : طبقه بندی سیارات براساس میزان شعاع مداری

```python
# pd.cut(): متغیرهای عددی را به گروه‌های دسته‌بندی شده تبدیل کنید و تحلیل‌های مختلفی بر روی داده‌های خود انجام دهید

# به مقادیر عددی "Orbital Radius (km)" تبدیل مقادیر رشته ای ستون 
planets['Orbital Radius (km)'] = pd.to_numeric(planets['Orbital Radius (km)'], errors='coerce')

# مرزهای دسته‌بندی به صورت عددی تعریف می‌شوند
bins = [0, 1000000, 10000000, 100000000, 1000000000, 10000000000]

#  pd.cut() با استفاده از "Orbital Radius (km)" دسته‌بندی ستون 
planets['Orbital Radius Category'] = pd.cut(planets['Orbital Radius (km)'], bins)

#  نمایش دیتافریم با ستون دسته‌بندی شده
planets[['Name', 'Orbital Radius (km)', 'Orbital Radius Category']]

# همین جا هم نتیجه گیری میشه کرد که سیارات خارجی که در بالا دیدیم شعاع مداری زیادی هم دارند
```

```python
# Orbital Radius Category حذف ستون 
planets.drop(columns='Orbital Radius Category')
```

# از کیلومتر به متر Orbital Radius (km) هدف : تبدیل مقادیر ستون 

```python
def km_to_meter(orbital_radius):
    
    if pd.notna(orbital_radius):
        return orbital_radius * 1000  # تبدیل کیلومتر به متر
    return orbital_radius

# apply با استفاد ه از متد 'Orbital Radius (km)' اعمال  تابع بالا به ستون 
planets['Orbital Radius (m)'] = planets['Orbital Radius (km)'].apply(km_to_meter)

#نهایی DataFrame چاپ 
planets
```

# Aggregate functions کاربرد

```python
result_min = planets['Discovery Date'].agg('min')  # حداقل مقدار
print(f"Minimum Discovery Date: {result_min}")

print('===========================================')

result_max = planets['Discovery Date'].agg('max') # حداکثر مقدار
print(f"Maximum Discovery Date: {result_max}")
print('===========================================')

result_mean = planets['Mean Distance to the Sun (km)'].agg('mean') # میانگین
print(f"Average Mean Distance to the Sun (km): {result_mean}")
print('===========================================')

result_nunique = planets['Discovery Date'].agg('nunique') # تعداد ردیف‌های غیر‌تکراری
print(f"Number of Unique Discovery Dates: {result_nunique}")
print('===========================================')

result_value_counts = planets['Discovery Date'].agg('value_counts') # توزیع فراوانی مقادیر
print(f"Value Counts for Discovery Dates:\n{result_value_counts}")
print('===========================================')

result_sum = planets['Mass (kg)'].agg('sum')
print('Sumation of Mass planets : ',result_sum)

print('===========================================')

result_multiple = planets['Discovery Date'].agg(['min', 'max', 'mean', 'nunique']) # استفاده از چند تابع به صورت همزمان
print(f"Summary for Discovery Dates:\n{result_multiple}")
```
