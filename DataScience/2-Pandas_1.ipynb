# 3- NumPy ساخت دیتافریم از آرایه های

# تعویض مقادیر یا به اصلاح جایگزین کردن یا آپدیت کردن

```python
data = [['Alex', 10], ['Bob', 12], ['Clarke', 13]] # الان 3 تا اسم داریم با 3 تا عدد 
df_list = pd.DataFrame(data=data, columns=['Name', 'Age'])
print(df_list)

print()

new_data = [['Reza', 22], ['Ali', 21], ['Frank', 22],['David', 25]]  # این جا 4 تا اسم و عدد داریم 
new_df_list = pd.DataFrame(data=new_data, columns=['Name', 'Age'])
print(new_df_list)

print()

df_list.update(new_df_list) # الان این جا ، اسم و سن دومی در اولی قرار میگیره
print('Update values : \n',df_list)

# اما نکته مهم این است که در دیتافریم اصلی فقط مقادیری که قبلاً وجود داشتند آپدیت می‌شوند و مقادیر جدیدی که به دیتافریم جدید اضافه شده است، به دیتافریم اصلی اضافه نمی‌شوند
# به عبارت دیگر، تنها مقادیر موجود در دیتافریم اصلی که با مقادیر جدید همخوانی دارند، جایگزین می‌شوند یعنی 3 تای اولش اضافه می شوند در این مثال

print('============================================================')

# مثال دوم

df_dict = pd.DataFrame({'A': [1, 2, 3],
                        'B': [400, 500, 600]})
print(df_dict)

print()

new_df_dict = pd.DataFrame({'B': [4, 5, 6],
                            'C': [7, 8, 9]})
print(new_df_dict)

print()

df_dict.update(new_df_dict) # مقادیرش آپدیت شد در جدول اول B الان توی این یکی،  ستون 
print('Update values : \n',df_dict)
```

# 2- نمایش سطر ها

```python
# This code uses the iterrows() function in Pandas to iterate through all the rows of a DataFrame, returning pairs (index, Series) in order.
# Each pair includes the details of an index and the corresponding row as a Series.

# نکته مهم این است که این تابع همراه با مشخصات یک ایندکس نیز اطلاعات را باز می‌گرداند

list(event.iterrows())  # ولی مشکلش این هست که کل فایل رو میخونه و همه ردیف ها رو مینویسه  و پارامتری نمی گیره و مثل ستون ها نیست رفتارش

# ولی ما معمولا با ستون ها کار داریم مثلا همه اون هایی که این ویژگی رو دارند انتخاب بشوند

# هست head(),tail(),sample() بهترین راه حلش همون 
```

```python
# loc: : برای انتخاب سطرها (ردیف‌ها) با استفاده از نام برچسب (نام ردیف) عمل می‌کند
# iloc: برای انتخاب سطرها (ردیف‌ها) با استفاده از شماره مکانی (اندیس) عمل می‌کند

# ix : بود مثالی ننوشتم و بررسی نکردم iloc و loc به دلیل حذف شدن در نسخه های جدید پانداس چون که ابهام ایجاد می کرد جون ترکیبی از 
```

```python
# location ===> loc
# integer location ===> iloc

event.iloc[0] # برای نمایش داده‌های مربوط به ردیف اول  از دیتافریم
```

```python
event.iloc[:, 0] # یعنی از همه سطر ها ستون 0 اون ها رو انتخاب کرده ، اون کاما سطر و ستون رو جدا می کنه

# توی براکت برای پارمتر اول که برای سطر ها هست : نوشته بعدش هم اسم ستون هارو وارد کرده ، این کار یعنی عملا انتخاب کل اون ستون ها iloc , loc پس یک نکته : هرگاه دیدی که بعد
```

```python
event.iloc[1:3, 0] # ازایندکس شماره 1 شروع میشه و تا 3 ولی خود 3 حساب نمیشه پس تا دومین ایندکس که با احتساب شماره 0 میشه سومین خانه  ،  وستون 0 آن ها مثل بالایی
```

```python
event.iloc[[0, 1, 5], 0] # یعنی فقط این شماره ایندکس ها باز هم ستون 0 آن ها 
```

```python
event.iloc[1, 6]  # است OS یعنی سطر دوم چون از 0 حساب کنیم میشه دوم  ، ستون ششم که مربوط به
```

```python
event.loc[1, 'OS']  # یعنی همون بالایی که نوشتم event.iloc[1, 6] ولی همین رو بالا باید مینوشتیم OS الان یعنی سطر دوم و ستون 
```

```python
selected_columns_loc = event.loc[:, ['displayId', 'timestamp', 'pageId', 'widgetId']] # یعنی همه سطر ها ولی فقط این 4 تا ستونشون رو میخواهیم
print(selected_columns_loc)
```

```python
selected_columns_iloc = event.iloc[:, [0, 1, 2, 3]]  # نوشته شده iloc همون بالایی هست که با 
print(selected_columns_iloc)
```

# پیدا کردن مقادیر تکراری

```python
import pandas as pd

# ساخت یک DataFrame نمونه
data = {'Name': ['John', 'Anna', 'John', 'Anna', 'John', 'Anna'],
        'Age': [25, 30, 25, 30, 25, 30]}
df = pd.DataFrame(data)

# یافتن مقادیر تکراری بر اساس تمام ستون‌ها
duplicates = df[df.duplicated()]

print("Duplicate rows:")
duplicates
```

# پاک کردن مقادیر تکراری

```python
import pandas as pd

# ساخت یک DataFrame نمونه
data = {'Name': ['John', 'Anna', 'John', 'Anna', 'John', 'Anna'],
        'Age': [25, 30, 25, 30, 25, 30]}
df = pd.DataFrame(data)

# حذف مقادیر تکراری بر اساس تمام ستون‌ها
df_unique = df.drop_duplicates()

print("DataFrame after removing duplicates:")
df_unique
```
